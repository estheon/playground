Lists are iterables that can contain practically any data type.

```python
# Create list
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
vegetables = list(("carrot","celery","broccoli"))
# Add an individual element
fruits.append("pear")
print(fruits)
# Add multiple elements
fruits.extend(["starfruit","papaya","orange"])
print(fruits)
# Add an element at a specific position
fruits.insert(1,"bread")
print(fruits)
# Get the index of the first element that matches the string
print(vegetables.index("broccoli"))
```

There's several ways to delete elements from lists.

```python
# Delete an element (bread is not a fruit) using pop(), which keeps the value
outlier = fruits.pop(1)
print(fruits)
print(outlier)
# Delete an element using del , which discards the value altogether
    # I used index() within [] b/c I was too lazy to look up the specific index of "kiwi"
del fruits[fruits.index("kiwi")]
print(fruits)
# Delete elements using an empty list and slicing (using positive numbers, not negative numbers)
fruits[5:7] = []
print(fruits)
# Sort in alphabetical order
vegetables.sort()
print(vegetables)
# Add another vegetable, sort in reverse order
vegetables.append("garlic")
vegetables.sort(reverse=True)
print(vegetables)
```

You can make nested lists:

```python
L = ['a', 'b', ['cc', 'dd', ['eee', 'fff']], 'g', 'h']
print("Using positive indexing:")
print(L[2])
# Prints ['cc', 'dd', ['eee', 'fff']]

print(L[2][2])
# Prints ['eee', 'fff']

print(L[2][2][0])
# Prints eee

print("Using negative indexing")
print(L[-3])
# Prints ['cc', 'dd', ['eee', 'fff']]

print(L[-3][-1])
# Prints ['eee', 'fff']

print(L[-3][-1][-2])
# Prints eee
```

When one is manipulating values within a nested list, the methods must be used on the nested list using the index:

```python
simpleL = ['a', ['bb', 'cc'], 'd']
simpleL[1][1] = 0
print(simpleL)
# Prints ['a', ['bb', 0], 'd']
```

List comprehension is a technique used for condensing code involving list creation and `for` loops and `if` statements.  
Multiple `for` loops / `if` statements can be used at one time.

```python
fruats = [x for x in fruits if "a" in x]
print(fruats)
picky = [x for x in fruits if x == "mango"]
print(picky)
compsci = [x+y for x in ['Python ','C '] for y in ['Language','Programming']]
print(compsci)
```
